Jest bardzo dużo kolumn ( aż 29 ). Na początku, należy wyodrębnić cechy nieistotne w kontekście tworzenia modelu ML, by ułatwić sobie dalszą analizę i zredukować wymiar pracy.
Jeśli będziemy dokonywać uczenia nadzorowanego nad jedną z trzech cech: 'Liczba pobran', 'Liczba wyswietlen', 'Stopien otwartosci'. Nieistonymi kolumnami będą prawdopodobnie:
- Id zasobu
- Link zasobu
- Nazwa
- Link zbioru
- Link instytucji
- Tytul
- Data utworzenia instytucji

Zredukowano ramkę danych o 7 cech

```python
columns_to_drop = ['Id zasobu','Link zasobu','Nazwa','Link zbioru','Link instytucji','Tytul','Data utworzenia instytucji']
df = df.drop(columns_to_drop,axis=1)
```

Poza kolumną 'Formaty po konwersji', mamy bardzo mało brakujących danych ( mnież niż 1 % ). Więc nie należy się przejmować powodami ich nieobecności. Możemy skategoryzować je jako missing completely at random ( MCAR ). Następnie usunąć rekordy z brakującymi danymi.

```python
df.isna().mean()
```

```python
df = df.dropna(subset=['Typ', 'Format', 'Data utworzenia zasobu',
       'Data modyfikacji zasobu', 'Stopien otwartosci',
       'Zasob posiada dane wysokiej wartosci', 'Zasob posiada dane dynamiczne',
       'Zasob posiada dane badawcze', 'Zasob zawiera wykaz chronionych danych',
       'Liczba wyswietlen', 'Liczba pobran', 'Id zbioru danych',
       'Zbior danych posiada dane wysokiej wartosci',
       'Zbior danych posiada dane dynamiczne',
       'Zbior danych posiada dane badawcze', 'Data utworzenia zbioru danych'])
```

Widać, że większość zasobów, pozostała w takim samym oryginalnym formacie

```python
df.isna().mean()
```

Kolumnę 'Formaty po konwersji' uznamy jako missing not at random  ( MNAR ) ponieważ, brakujące dane są spowodowane brakiem konwersji danych -> czyli istnieje jasno widoczny mechanizm przyczynowo skutkowy.
Należy podmienić wartości NaN na wartości 'unchanged'.
Jedynym problemem, jest że niektóre 'Formaty po konwersji' są nie-atomowe, mają dwie lub więcej wartości np: 'csv, json-ld'.

Podmiemiono na stałe wartości NaN w kolumnie 'Formaty po konwersji' na wartość unchanged

```python
df = df.fillna("unchanged")
df.head(5)
```

Otrzymano ramkę danych bez brakujących danych

```python
df.isna().mean()
```

Właściwie to Format po Konwersji, będzie teraz aklualnym Formatem, więc można podmienić Wartości Format, tymi aktulanymi. Oczywiście pomijając wartości 'unchanged'
Teraz można pozbyć się kolumny 'Format po Konwersji'. Ponownie redukując liczbę cech

```python
df['Format'] = df.apply(lambda row: row['Formaty po konwersji'] if row['Formaty po konwersji'] != 'unchanged' else row['Format'], axis=1)
```

```python
df = df.drop('Formaty po konwersji',axis=1)
df.head(5)
```

Zbadajmy kardynalność cech kategorycznych:
 - najpierw należy sprawdzić czy wszystkie kolumny mają poprawny typ


```python
df.info()
```

Zmienne określające Daty, mają obecnie typ object, należy zamienić go na typ datetime

Zmienne określające Id, są zapisane jako int64 lub float64. Stopien otwartosci jako float64. Należy zamienić je na typ str, żeby dalej zostały uznane jako zmienna kategoryczna

```python
df["Data utworzenia zasobu"] = pd.to_datetime(df["Data utworzenia zasobu"])
df["Data modyfikacji zasobu"] = pd.to_datetime(df["Data modyfikacji zasobu"])
df["Data utworzenia zbioru danych"] = pd.to_datetime(df["Data utworzenia zbioru danych"])
df["Data modyfikacji zbioru danych"] = pd.to_datetime(df["Data modyfikacji zbioru danych"])
```

```python
df["Stopien otwartosci"] = df["Stopien otwartosci"].astype(str)
df["Id zbioru danych"] = df["Id zbioru danych"].astype(str)
df["Id instytucji"] = df["Id instytucji"].astype(str)
```

Wygląda na to, że większość zmiennych posiada niską kardynalność cech.
Właściwie tylko Id's, co było do przewidzenia mają wysoką, oraz Format.

```python
categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
for column_name in categorical_columns:
    unique_values = len(df[f'{column_name}'].unique())
    # 10 jako granica mocy zbioru pomiędzy wysoką a niską
    cardinality = 'wysoka' if unique_values > 10 else 'niska'
    print(f'\nLiczba etykiet zmiennej {column_name}: {unique_values}, Moc zbioru: {cardinality}')
```

Możemy teraz zredukować kardynalność cech zmiennej Format, pozbywając się nie-atomowych etykiet. Ponieważ, w kolumnie powinny znajdować się dane niepodzielne, jest to też ważne w kontekście uczenia maszynowego, pozbędziemy się wieloznaczności informacji, co powinno pozytywnie wpłynąć na nauczanie.

Ilość nie-atomowych etykiet stanowi zaledwie mniej niż jeden procent naszych danych, więc redukcja ich do wartości 'csv', nie powinna być szkodliwa.

Zredukowano kardynalność cechy z 29 do 26

```python
nie_atomowe = [string for string in df["Format"].unique() if ',' in string]
print(nie_atomowe)
```

```python
print(round(len(df[df['Format'].isin(nie_atomowe)]) / len(df),3),"%")
```

```python
df['Format'] = df.apply(lambda row: 'csv' if row['Format'] in nie_atomowe else row['Format'], axis=1)
```

```python
unique_values = len(df['Format'].unique())
cardinality = 'wysoka' if unique_values > 10 else 'niska'
print(f'\nLiczba etykiet zmiennej Format: {unique_values}, Moc zbioru: {cardinality}')
```
